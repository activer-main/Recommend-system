## 匯入檔案 & 獲取活動資料

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import requests
import json
import csv
import numpy as np
import pandas as pd

# get activit
params = {
    'SortBy': 'CreatedAt',
    'OrderBy': 'ascending',
    'Page': 1,
    'CountPerPage': 360
}

url = 'http://danny10132024-andes.nord:5070/api/Activity?SortBy={}&OrderBy={}&Page={}&CountPerPage={}'.format(params['SortBy'], params['OrderBy'], params['Page'], params['CountPerPage'])

response = requests.get(url)

# 未接收到 API
if response.status_code != 200:
    print("Failed to get data from API.")
    exit(0)

data = json.loads(response.content)['searchData']

len(data)
```
