```python
import pandas as pd

# 创建示例 DataFrame
row_activity = pd.DataFrame({'id': [1, 2, 3],
                             'tags': ['tag1|tag2', 'tag2|tag3', 'tag1']})

# 将 "tags" 列的值转换为 Dummies DataFrame
tags_dummies = row_activity['tags'].str.get_dummies('|')

# 创建新的空 DataFrame，其中包含与 "tags_dummies" 相同的列
new_df = pd.DataFrame(columns=tags_dummies.columns)

# 将 "tags_dummies" DataFrame 与 "new_df" DataFrame 连接
new_df = pd.concat([new_df, tags_dummies], ignore_index=True)

# 输出新的 DataFrame
new_df
```
