```python
import re

text = "這是一段包含標點符號、數字、空格和特殊符號！123abc#$%\n"

# 去除標點符號、數字、特殊符號，以及轉換為小寫
processed_text = re.sub(r"[^\u4e00-\u9fa5a-zA-Z\s\n\f]", "", text).lower()
processed_text = processed_text.replace('\n', '')


processed_text  # "這是一段包含標點符號數字空格和特殊符號abc"
```

```python
import pandas as pd
import base64
from bs4 import BeautifulSoup

# Read CSV file
row_activity = pd.read_csv('data/raw/activity.csv')

# Get encoded content
encode_content = row_activity['content']

# Decode Base64 to UTF-8
decode_content = encode_content.apply(lambda x: base64.b64decode(x).decode('utf-8'))

# Extract text from HTML
text_content = decode_content.apply(lambda x: BeautifulSoup(x, 'html.parser').get_text().strip())

# 去除標點符號、數字、特殊符號，以及轉換為小寫
format_text = text_content.apply(lambda x: re.sub(r"[^\u4e00-\u9fa5a-zA-Z]", "", x).lower())

# 去除 \n\f\s
format_text = format_text.apply(lambda x: x.replace('\n', '').replace('\xa0', '').replace('\r', ''))

row_activity['content'] = format_text

format_text
```

```python
import pandas as pd

# 创建示例 DataFrame
row_activity = pd.DataFrame({'id': [1, 2, 3],
                             'tags': ['tag1|tag2', 'tag2|tag3', 'tag1']})

# 将 "tags" 列的值转换为 Dummies DataFrame
tags_dummies = row_activity['tags'].str.get_dummies('|')

# 创建新的空 DataFrame，其中包含与 "tags_dummies" 相同的列
new_df = pd.DataFrame(columns=tags_dummies.columns)

# 将 "tags_dummies" DataFrame 与 "new_df" DataFrame 连接
new_df = pd.concat([new_df, tags_dummies], ignore_index=True)

# 输出新的 DataFrame
new_df
```
